# Amazon Reviews Scraping

## Create Amazon Reviews Spider

```python
import scrapy
import pandas as pd
from scrapy.crawler import CrawlerProcess

class Amazon_Review_Spider(scrapy.Spider):
    name='amazon_review_spider'
    def start_requests(self):
        # The page to crawl
        urls='https://www.amazon.com/Razer-Hammerhead-Wireless-Bluetooth-Earbuds/product-reviews/B07Z77V1RH/ref=cm_cr_arp_d_paging_btm_next_2?ie=UTF8&reviewerType=all_reviews&pageNumber='
        for x in range(1,26): #the number of pages to crawl
            u=urls+str(x)
            yield scrapy.Request(url=u,callback=self.parse,headers= {'User-Agent': 'Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/62.0.3202.94 Safari/537.36'})
            
    def parse(self,response):
        review_title='//a[@data-hook="review-title"]/span/text()'
        review_star='//i[@data-hook="review-star-rating"]/span/text()'
        for x in range(1,11):
            review_content='string(/html/body/div[1]/div[2]/div[1]/div[1]/div/div[1]/div[5]/div[3]/div['+str(x)+']/div/div/div[4]/span/span)'
            dt['review content']=dt['review content']+response.xpath(review_content).extract()
        List1=[]
        dt['review title']=dt['review title']+response.xpath(review_title).extract()
        star=response.xpath(review_star).extract()
        List2=[]
        for x in star:
            List2.append(int(x[0]))
        dt['review rating']=dt['review rating']+List2

dt={}
dt['review rating']=[]
dt['review title']=[]
dt['review content']=[]
process = CrawlerProcess()
process.crawl(Amazon_Review_Spider)
process.start()
dt=pd.DataFrame(dt)
dt.to_csv("Amazon_Review.csv") #filename
```
